```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#load dataset
epl_df = pd.read_csv('E:\\Project Python\\EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
#Create 2 new columns
epl_df['MinsPerMatch']=(epl_df['Mins'] /epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch']=(epl_df['Goals'] /epl_df['Matches']).astype(float)
epl_df.head()
```

```python
#Total Goals
Total_Goals = epl_df['Goals'].sum()
print(Total_Goals)
```

```python
#Penalty Goals
Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print(Total_Penalty_Goals)
```

```python
#Penalty Attempts
Total_Penalty_Attempts = epl_df['Penalty_Attempted'].sum()
print(Total_Penalty_Attempts)
```

```python
#pie chart for Penalties missed vs scored
plt.figure(figsize=(13, 6))
pl_not_scored = epl_df['Penalty_Attempted'].sum() - Total_Penalty_Goals
data = [pl_not_scored,Total_Penalty_Goals]
labels=['Penalties Missed','Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels ,colors = color,autopct ='%1.1f%%',explode = [0.1,0.1],shadow = True)

plt.axis('equal')
plt.show()
```

```python
#Club With Maxmimum Players in their squad
epl_df['Club'].value_counts().nlargest(5).plot(kind = 'bar',figsize=(12, 6) ,color = sns.color_palette('viridis'))
```

```python
#Club With least Players in their squad
epl_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar',figsize=(12, 6) ,color = sns.color_palette('viridis'))
```

```python
#Players Based on age groups
Under20 = epl_df[epl_df['Age'] <= 20 ]
Age20_25 = epl_df[(epl_df['Age'] > 20 ) & (epl_df['Age'] <= 25)]
Age25_30 = epl_df[(epl_df['Age'] > 25 ) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age'] > 30 ]
```

```python
x = np.array([Under20['Name'].count(),Age20_25['Name'].count(),Age25_30['Name'].count(),Above30['Name'].count()])
My_Labels = ["<=20", ">20 & <=25", " >25 & 30"," >30 "]
plt.title('Total Players With Age Groups',fontsize = 20)
plt.pie(x, labels = My_Labels ,autopct ='%1.1f%%' ,shadow = True)
plt.show()
```

```python
#Total Under 20 Players in each Club
Players_Under_20 = epl_df[epl_df['Age'] < 20 ]
Players_Under_20['Club'].value_counts().plot(kind = 'bar',color = sns.color_palette('viridis'))
```

```python
#Average age of Players in each Club
plt.figure(figsize=(12, 6))
sns.boxplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
```

```python
num_Player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/ num_Player
data.sort_values(ascending = False)
```

```python
#Total assists from each Club
Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_style("whitegrid")
color_codes=True
ax = sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by="Assists"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation = 75)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('plot of Club vs Total Assists',fontsize = 20)
```

```python
#Total Passes_Attempted from each Club
Passes_by_clubs = pd.DataFrame(epl_df.groupby('Club',as_index=False)['Passes_Attempted'].sum())
sns.set_style("whitegrid")
color_codes=True
ax = sns.barplot(x='Club',y='Passes_Attempted',data=Passes_by_clubs.sort_values(by="Passes_Attempted"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Passes_Attempted",fontsize=20)
plt.xticks(rotation = 50)
plt.rcParams["figure.figsize"] = (20,8)
plt.title('Plot of Club vs Total Passes Attempted',fontsize = 30)
```

```python
#Top 10  Players With Most Passes
epl_Passes=epl_df.sort_values(by ='Passes_Attempted',ascending = False )[:10]
plt.figure(figsize=(20, 6))
plt.title('Players With The Most Passes',fontsize = 20)
c = sns.barplot(x=epl_Passes['Name'],y= epl_Passes['Passes_Attempted'],label='Players' , color = 'blue')
plt.ylabel('Number Of Passes')
c.set_xticklabels(c.get_xticklabels(),rotation = 45)
c
```

```python
#pie chart Goals with Assists and without Assists
plt.figure(figsize=(14, 7))
Assists = epl_df['Assists'].sum()
data = [Total_Goals - Assists, Assists]
labels=[' Goals Without Assists ',' Goals Assists']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels ,colors = color,autopct ='%1.1f%%',explode = [0.1,0.1],shadow = True)

plt.axis('equal')
plt.show()
```
